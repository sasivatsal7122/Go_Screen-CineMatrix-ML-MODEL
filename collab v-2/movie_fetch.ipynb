```python
import pandas as pd
import requests
import difflib
```

```python
mymdf = pd.read_csv("moviesss.csv")
mymdf.head(10)
```

```python
tmdb = pd.read_csv("tmdb_5000_movies.csv")
tmdb.head(10)
```

```python
titles=[]
movie_ls = ['Iron Man 2 (2010)',
'Iron Man 3 (2013)',
'Avengers, The (2012)',
'Avengers: Infinity War - Part I (2018)',
'Captain America: Civil War (2016)',
'Incredible Hulk, The (2008)',
'Ant-Man (2015)',
'Avengers: Age of Ultron (2015)',
'Heart and Souls (1993)',
'Untitled Spider-Man Reboot (2017)']
for movie in movie_ls:
    close_match = difflib.get_close_matches(movie, list(tmdb['title']))
    if bool(close_match):
        titles.append(close_match[0])
print(titles)  
```

```python
movie_id_ls=[]
for title in titles:
    blah = tmdb.loc[tmdb['original_title'] == title]
    movie_id = int(blah['id'])
    movie_id_ls.append(movie_id)
print(movie_id_ls)
```

```python
def fetch_poster(movie_id):
    url = "https://api.themoviedb.org/3/movie/{}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US".format(movie_id)
    data = requests.get(url)
    data = data.json()
    poster_path = data['poster_path']
    full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    print(full_path)
```

```python
for m_id in movie_id_ls:
    (fetch_poster(m_id))
```
